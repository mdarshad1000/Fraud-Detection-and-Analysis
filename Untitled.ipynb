# Detection and Analysis of Credit Card Fraud

## Introduction
<p>When we make any transaction while purchasing any product online, a good amount of people prefer credit cards. A high credit limit in credit cards sometimes helps us making expensive purchases even if we don’t have the savings at that time. But, on the other hand, these features are misused by cyber attackers.
In order to tackle such a problem, a system needs to be built which can classify such transactions based on its characteristics and mark them fraudulent for the bank officials to act upon.
In this study, we will try to build an algorithm to do the same, while using a real dataset consisting transactional data of European Cardholders. The data was recorded within a span of 2 days in the September of 2013. Due to confidentiality issues, the features obtained were transformed and named in as V1, V2, ... , V28. Other background information about the customers was also removed.</p>

 <b>Under sampling</b>
 <p>Build a sample dataset containing similar distribution of normal tansaction and Fradulent transaction</p>

```python
 legit_sample =  legit.sample(n=492)
```

<b>Concatenating two DataFrames</b>

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

The mean values of Fraud and Legit transactions in the new dataset vs the old dataset are very similar. This indicates that we have a good sample.

<b>Splitting the Data into features and Targets</b>

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

<b>Splitting the Data into Train and Test</b>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

<h2>Model Training</h2>

<h3>1. Logistic Regression</h3>

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(X_train, Y_train)
```

 <b>Accuracy Score</b>

```python
from sklearn.metrics import accuracy_score
```

Training Data

```python
logReg_prediction_train = logistic_regression.predict(X_train)
```

```python
logReg_accuracy_train = accuracy_score(logReg_prediction_train, Y_train)
```

```python
print("Training accuracy score for logistic Regression model is", logReg_accuracy_train)
```

Test Data

```python
logReg_prediction_test = logistic_regression.predict(X_test)
```

```python
logReg_accuracy_test = accuracy_score(logReg_prediction_test, Y_test)
```

```python
print("Test accuracy score for logistic Regression model is", logReg_accuracy_test)
```

Training data and Test data accuracy are very comparable, that implies that the model has not Overfitted or Underfitted
